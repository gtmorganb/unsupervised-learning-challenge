```python
# Initial imports
import pandas as pd
from pathlib import Path
```

```python
 # Data loading
file_path = Path("crypto_data.csv")
crypto_df = pd.read_csv(file_path)
crypto_df.head()
```

```python
crypto_rows = len(crypto_df)
crypto_rows
```

```python
#drop non-trading currencies
crypto_trade = crypto_df[crypto_df['IsTrading'] != 0]
crypto_trade
```

```python
#drop IsTrading from dataset 
crypto_new = crypto_trade.drop(['IsTrading'], axis = 1)
crypto_new
```

```python
#dropping rows with NaN
crypto_clean = crypto_new.dropna(axis=0, how='any')
crypto_clean
```

```python
#finding mined values greater than 0 
crypto_cleaned = crypto_clean[crypto_clean['TotalCoinsMined'] > 0]
crypto_cleaned
```

```python
#remove Coin Name from dataframe
crypto_next = crypto_cleaned.drop(['CoinName'], axis = 1)
crypto_next
```

```python
#renaming columns and setting index
crypto_next.columns = ['CryptoID', 'Algorithm', 'ProofType', 'TotalCoinsMined', 'TotalCoinSupply']

crypt_final =crypto_next.set_index("CryptoID")
crypt_final
```

```python
#changing Total Coin Supply to float before get dummies since it is numerical 
crypt_final["TotalCoinSupply"] = crypt_final.TotalCoinSupply.astype(float)
list(crypt_final.dtypes)
```

```python
crypto_dum = pd.get_dummies(crypt_final)
crypto_dum
```

```python
#additional imports for dummy variables and standardizing data
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
crypto_scale = scaler.fit_transform(crypto_dum)
crypto_scale
```

# Dimensionality Reduction

```python
pca = PCA(n_components=0.9)
```

```python
crypto_pca = pca.fit_transform(crypto_scale)
```

```python
 # Transform PCA data to a DataFrame
pca_df = pd.DataFrame(data=crypto_pca)
pca_df
```

```python
 # Fetch the explained variance
pca.explained_variance_ratio_
```

```python
#more dependencies
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
#tSNE on pca data
tsne = TSNE(learning_rate=100)
transformed_features = tsne.fit_transform(crypto_pca)
```

```python
 # Check that dimensions have been reduced to two
transformed_features.shape
```

```python
crypto_dum['x'] = transformed_features[:,0]
crypto_dum['y'] = transformed_features[:,1]
```

```python
plt.scatter(crypto_dum['x'], crypto_dum['y'])
plt.show()
```

# Cluster Analysis with k-Means

```python
#Define the inertia and K cluster
inertia = []
k = list(range(1, 10)) 

# Calculate the inertia for the range of K values
for i in k:
   km = KMeans(n_clusters=i, random_state=0)
   km.fit(pca_df)
   inertia.append(km.inertia_)
```

```python
# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
 # Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for customer data')
plt.show()
```

```python
def get_clusters(k, data):
    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Train the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
clusters = get_clusters(6, crypto_dum)
clusters
```
